```python
import os
import shutil
import pandas as pd

from tqdm.auto import tqdm
from google.colab import drive

drive.mount('/content/gdrive', force_remount=True)
```

```python
src_dir = 'ssv2'
dst_dir = 'ssv2_mini'

sample_sizes = {
    'train': 20,
    'test': 3,
}

for dstype in ['train', 'test']:
    tqdm.write(f"Copying '{dstype}' from {src_dir} to {dst_dir}")
    pb1 = tqdm(total=174, desc=f"Classes", position=0)
    n = sample_sizes[dstype]
    pb2 = tqdm(total=n, desc='Files', position=1)

    for src, _, files in os.walk(os.path.join(src_dir, dstype)):
        if not files:
            continue
        pb2.reset()

        label = os.path.basename(src)
        dst = os.path.join(dst_dir, dstype, label)
        if not os.path.exists(dst):
            os.makedirs(dst)
        
        for i, fname in enumerate(files):
            shutil.copy2(os.path.join(src, fname), dst)
            pb2.update(1)
            if i+1 == n:
                break
        else:
            tqdm.write(f'Warning - class {label} has only {i+1} {dstype} samples')
        
        pb2.refresh()
        pb1.update(1)
    
    pb1.refresh()
    pb1.close()
    pb2.close()
```

```python
def make_list(root, dstype):
    data = []
    for path, _, files in os.walk(os.path.join(root, 'raw', dstype)):
        if not files:
            continue
        
        for file in files:
            id, ext = os.path.splitext(file)
            if ext == '.webm':
                label = os.path.basename(path)
                data.append((id, label))
    
    csv_path = os.path.join(root, f'{dstype}_list.csv')
    pd.DataFrame(data, columns=['id', 'label']).to_csv(csv_path, index=False)

make_list('ssv2_mini', 'train')
make_list('ssv2_mini', 'test')
```
